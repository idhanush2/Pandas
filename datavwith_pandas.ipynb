```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cancer_df = pd.read_csv('cancer.csv')
```

```python
cancer_df.head()
```

```python
cancer_df.tail()
```

```python
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'target', data = cancer_df)
```

```python
sns.scatterplot(x = 'mean radius', y = 'mean area', hue = 'target', data = cancer_df)
```

```python
sns.pairplot(cancer_df, hue = 'target', vars = ['mean radius','mean texture','mean area','mean perimeter','mean smoothness'])
```

```python
sns.heatmap(cancer_df.corr())    #correlation
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(cancer_df.corr(), annot = True) #actual correlation numbers with color code
```

```python
sns.distplot(cancer_df['mean radius'], bins = 5, color = 'blue')
```
