```python
daily_return_df = pd.read_csv('crypto_daily_returns.csv')
daily_return_df
```

```python
mu = daily_return_df['BTC'].mean()     #avg
sigma = daily_return_df['BTC'].std()   #sd

plt.figure(figsize=(12,6))
daily_return_df['BTC'].plot.hist(bins=50, alpha = 0.5)     #plots histogram, aplha reduces transparency 
plt.grid()
plt.title('Histogram: mu=' + str(mu) + ',Sigma ='+ str(sigma))
```

```python
mu = daily_return_df['ETH'].mean()     
sigma = daily_return_df['ETH'].std()   

plt.figure(figsize=(12,6))
daily_return_df['ETH'].plot.hist(bins=30, alpha = 0.5, color = 'red')     
plt.grid()
plt.title('Histogram: mu=' + str(mu) + ',Sigma ='+ str(sigma))
```
